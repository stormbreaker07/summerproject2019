```python
import folium
import pandas

```

```python
data=pandas.read_csv("hospital_directory.csv")
```

```python
data["coordinates"].fillna(method ='ffill', inplace = True)  
lat=list(data["coordinates"])
lat[203]
```

```python
data["coordinates"].fillna(method ='ffill', inplace = True)  
lat=list(data["coordinates"])
address=list(data['Address_Original_First_Line'])
```

```python
map=folium.Map(location=[28.5976267,77.1083615],tiles="OpenStreetMap")
fg=folium.FeatureGroup(name="my map")

lat=list(data["coordinates"])

for i in range(0,100):
    print(i)
    lis=[]
    lis=(lat[i]).split(sep=",")
    fg.add_child(folium.Marker(location=[float(lis[0]),float(lis[1])],popup=str(address[i]),icon=folium.Icon(color="blue")))
map.add_child(fg)

map.save("map1.html")
```
